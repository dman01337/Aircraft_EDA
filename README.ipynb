

# Aircraft Risk Assessment Analysis



### Analysis Overview



The overall goal of this analysis is to evaluate the [Aviation Accident Database](https://www.kaggle.com/datasets/khsamaha/aviation-accident-database-synopses) to determine which aircraft poses the least amount of risk for injuries, while also gauging the best financial option. This analysis will be able to show which aircrafts should be taken into account for potential use.

### Business Goal



A company is expanding into the aircraft industry to
diversify its portfolio. Specifically, they are interested
in purchasing and operating airplanes for commercial
and private enterprises, but are unaware of the potential risks of aircraft. The goal here is to evaluate which aircraft provides the best fit for the company by assessing the personal injury risk of purchasing said aircraft.


### Understanding the Data



The [Aviation Accident Database](https://www.kaggle.com/datasets/khsamaha/aviation-accident-database-synopses) provides extensive data of aircraft accidents since the year 1948. We can use this dataset to find which aircrafts have the largest potential for injury. The dataset also includes a wealth of data related to aircrafts and the accidents over the past few decades to help us narrow down the best fits.

### Methods


The analysis extensively sifts through the database to compare the trends over the years of the injuries to the different types of aircrafts. This helps to determine which aircrafts would be a safer choice for the company to use in their business venture. We converted the data into tables showing the amount of fatalities per aircraft with the focus being the size of the aircraft.

## Conclusions


The analysis led us to three recommendations for the company when it comes to the aquisition of planes for commercial use.

- Firstly, we would not recommend the use of small class planes because their fatality rate is much higher than the larger sized planes. If the company is admamant about purchasing smaller planes then we would recommend the Cessna 180, Cessna 150, or the Cessna 172 because they have the lowest and most similar fatality rate.

- Large capacity planes would be ideal for commercial use since they can hold an ample number of people while also having fantastic safety ratings depending on the manufacturer. Boeing stands out from the data as having the lowest fatality rate among the large sized crafts. With that we would recommend the Boeing 777, Boeing 777-223, or the Boeing 777-222 models as being great choices for commercial use.

- Medium sized planes could also prove valuable assets. While they seat less people, they also are smaller crafts so less fuel is required to power them. They also boast fatality rates similar to the larger plane class making them good options for commercial use. Boeing also sits at the lowest fatality rates among the medium sized crafts as well. Our three top choices would be the Boeing 757-223, Boeing 737, or the Boeing 737-200
